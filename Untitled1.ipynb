```python
#optimisation du portfolio

from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

#mu → rendement annuel attendu
mu = rendement annuel estimé de chaque action
#
#S → matrice de covariance (le risque)
--> S est une matrice carré NxN
#
Ce que ça fait
*   Calcule les rendements quotidiens
*   Mesure comment les actions bougent ensemble
*   Annualise le tout

```python
# calculer le rendement annualisé prévu et la matrice de covariance annualisée
#de l'échantillon des rendements quotidiens des actifs

# moynne rendement = means returns
mu = expected_returns.mean_historical_return(df)
S = risk_models.sample_cov(df)

```

```python
#optimiser le sharp ratio = rapport net
ef = EfficientFrontier(mu, S) #create the efficient frontier Object
weights = ef.max_sharpe()

cleaned_weights = ef.clean_weights()
print (cleaned_weights)

#verbose = True --> affiche les résultats de façon détaillée
ef.portfolio_performance(verbose=True)
```

```python
#obtenir la répartition discrète de chaque action par titre
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices

# Si j'investie 5000€
portfolio_val = 5000

# trasformer le pourcentages → en nombre d’actions réelles.
latest_prices = get_latest_prices(df)

#les pourcentages calculés par l’optimisation.
weights = cleaned_weights

#on donne a da qui est un objet les pourcentages, prix actuel de chaque action et mon budget
#pourcentage --> weights
# prix des actions --> latest_price
#budget --> portfolio_val
da = DiscreteAllocation(weights, latest_prices, total_portfolio_value=portfolio_val)

allocation, leftover =da.lp_portfolio()    # lp = linear programming |  leftover = argent qui reste

print ('Discrete allocation:', allocation)

print ('les fonds restants:', leftover)





```

```python
#creer une fonction pour recuperer les noms des entreprises
import yfinance as yf

def get_company_name(symbol):
    try:
        stock = yf.Ticker(symbol)
        return stock.info.get("longName", symbol)
    except:
        return symbol

company_name = []

for symbol in allocation:
    company_name.append(get_company_name(symbol))

company_name

```
